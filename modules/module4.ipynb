```python

#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
This script reads a CSV file containing car sales data into a pandas DataFrame.
It assumes the CSV file is located in a 'data' directory relative to the script's location
and is encoded in 'utf-8'.
"""
import pandas as pd

""" Read the CSV file into a DataFrame
Ensure the file path is correct relative to the script's location
If the file is in a different directory, adjust the path accordingly
For example, if the file is in a 'data' folder, use '../data/Car_Sales.csv'
Here, we assume the file is in the same directory as this script
If the file is encoded in a different format, specify the encoding parameter
For example, if the file is encoded in 'utf-8', use encoding='utf-8'
If the file is large, consider using chunksize to read it in smaller parts
For this example, we will read the entire file at once
""" 
df = pd.read_csv('../data/Car_Sales.csv', encoding='utf-8')

# Display the first 5 rows of the DataFrame
df.head(5)
```

```python
# Display the DataFrame's structure and data types
df.dtypes
```

```python
# Display the DataFrame's information, including non-null counts and data types
df.info()
```

```python
# display null values in the DataFrame
# This will show the count of null values in each column
df.isnull().sum()
```

```python
# calculate the percentage of null values in each column
# This will show the percentage of null values in each column
df.isnull().mean() * 100
```

```python
# replace null values with a specific value
df['Price ($)'].fillna(0, inplace=False)
```

```python
# count Cars by make
count_make=df['Make'].value_counts() # iterates over the 'Make' column and counts occurrences of each unique value
count_make
```

```python
# Basic bar plot of the count of cars by make
import matplotlib.pyplot as plt     
count_make.plot(kind='bar', figsize=(10, 5), color='green')
plt.title('Count of Cars by Make',fontsize=20)
plt.xlabel('Make')
plt.ylabel('Number of Cars')     
plt.xticks(rotation=45)
plt.show()  
```
